```python
import pickle as pkl
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole
```

```python
# --- Load the QM9 data from pickle file
with open('./qm9_xas.pkl', 'rb') as data:
    df = pkl.load(data)
```

```python
df
```

```python
# --- Remove any NaN items from dataframe
clean_df = df.dropna()
clean_df
```

```python
# --- Save new cleaned dataframe
clean_df.to_pickle('./raw/qm9_xas_clean.pkl')
```

```python
index = 100

smiles = df.iloc[index]['SMILES']
mol = Chem.MolFromSmiles(smiles)
mol
```

```python
for atom in mol.GetAtoms():
    print(atom.GetFormalCharge())
```
